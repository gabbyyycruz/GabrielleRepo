````python
# Summary:

# 7 columns:
    # columns to be left alone: firstname (1), lastname (2), birthday (4)
    # columns to be cleaned: birthmonth (3), birthyear (5), course (6), phonenumber (7)
# list names of each column:
    # firstname = "fname"
    # lastname = "lastname"
    # birthmonth = "bmonths"
    # birthday = "bdays"
    # birthyear = "newyears"
    # course = "courses"
    # phonenumber = "newnumbers"
````

````python
import csv
import re

inputfile = csv.reader(open('recweekdata.csv','r'))
````

````python
# COLUMN 1: FIRSTNAME
# logic/process: 
    # 1. adding all first names to a list for convenience when writing to CSV

inputfile = csv.reader(open('recweekdata.csv','r'))

fname = []

for row in inputfile:
    fname.append(row[0])

fname.pop(0)
print(fname)
````

````python
# COLUMN 2: LASTNAME
# logic/process: 
    # 1. adding all last names to a list for convenience when writing to CSV

inputfile = csv.reader(open('recweekdata.csv','r'))
    
lastname = []

for row in inputfile:
    lastname.append(row[1])
    
lastname.pop(0)
print(lastname)
````

````python
# COLUMN 3: BIRTHMONTH
# logic/process: 
    # 1. opening data with excel 
    # 2. using vlookup to replace wrong inputs (spreadsheet also committed to github)
    # 3. adding formatted data to a list
    
inputfile2 = csv.reader(open('birthmonthtable.csv','r'))

bmonths = []

for row in inputfile2:
    bmonths.append(row[1])
    
bmonths.remove('birthmonth (values only)')
````

````python
# COLUMN 4: BIRTHDAY
# logic/process: 
    # 1. adding all birth days to a list for convenience when writing to CSV
    
inputfile = csv.reader(open('recweekdata.csv','r'))
    
bdays = []

for row in inputfile:
    bdays.append(row[3])
    
bdays.remove("birthday")
````

````python
# COLUMN 5: BIRTHYEAR
# logic/process: 
    # 1. adding all birth years from the "birthyear" column to a list
    # 2. retrieving the 2 characters of birth year 
    # 3. adding a "19" or a "20" in front of the birth year as appropriate 
    # 4. adding all newly-formatted birth years to another list
    
inputfile = csv.reader(open('recweekdata.csv','r'))
    
year = []

for row in inputfile:
    year.append(row[4])

newyears = []

for y in year:
    if y == "birthyear":
        continue
    elif(int(y[-2:])<90):
        newyears.append("20"+y[-2:])
    else:
        newyears.append("19"+y[-2:])
````

````python
# COLUMN 6: COURSE
# logic/process: 
    # 1. adding all course inputs to a list
    # 2. getting a set of all possible inputs
    # 3. replacing course names as appropriate, removing "BS" in front of string
    # 4. adding all newly-formatted courses to another list, adding "BS" once again
    
inputfile = csv.reader(open('recweekdata.csv','r'))

courses0 = []

for row in inputfile:
    courses0.append(row[5])
    
courses0.remove("course")
courseset = set(courses0)

print(courseset)
````

````python
courses1 = []

for course in courses0:
    if course[0:2]=="BS":
        courses1.append(course[3:])
    else:
        courses1.append(course)

courses = []
        
for course in courses1:
    if course=="ITE":
        courses.append("BS Information Technology Entrepreneurship")
    elif course=="MGT":
        courses.append("BS Management")
    elif course=="CTM":
        courses.append("BS Communications Technology")
    elif course=="LM":
        courses.append("BS Legal Management")
    elif course=="ME":
        courses.append("BS Management Engineering")
    else:
        courses.append("BS "+course)
````

````python
# COLUMN 7: PHONENUMBER
# logic/process: 
    # 1. adding all phone numbers from the last column to a list
    # 2. retrieving the last 12 characters of each phone number to remove the "+63" 
    # 3. adding a "0" in front of each phone number
    # 4. adding all newly-formatted phone numbers to another list

inputfile = csv.reader(open('recweekdata.csv','r'))
    
numbers = []

for row in inputfile:
    numbers.append(row[6])

newnumbers = []

for number in numbers:
    if number == "phonenumber":
        continue
    else:
        newnumbers.append("0"+number[-12:])
````

````python
# combining all lists together through pandas module

import pandas

df = pandas.DataFrame(data={"firstname": fname, "lastname": lastname,"birthmonth": bmonths, "birthday": bdays,
                           "birthyear": newyears, "course": courses, "phonenumber": newnumbers})
df.to_csv("cleanrecweekdata.csv", sep=',',index=False)
````
